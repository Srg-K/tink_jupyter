```python
#создаем датафрейм

data = pd.DataFrame(
    {
        "col_1":['a', 'b', 'c', 'd', 'e', 'f', 'j', 'h', 'i', 'j'],
        "col_2":[777, 778, 779, 780, 781, 782, 783, 784, 785, 786],
        "some_value":list(2 * n ** 3 for n in range(10)),
        "another_value":list(3 * n ** 4 for n in range(10)),
        "another_one_value":list(-n ** 6 for n in range(10))[::-1],
        "dt":pd.date_range('1/1/2049', periods=10),
        "type":['cat' if i%2==0 else 'dog' for i in range(10)]
    },
    index=[x for x in range(10)])

data
```

### Группировка и аггрегация

```python
#среднее по группам
data.groupby('type').mean()

# median()
# min()
# max()
# sum()
# count()
```

```python
#выбираем нужные значения
#сбрасываем индекс
data.groupby('type')[['col_2', 'another_one_value']].mean().reset_index()
```

```python
#более гибкий способ агрегировать
data.groupby('type').agg({
    'col_2':'sum', #сумма
    'some_value':'mean', #среднее
    'another_value': lambda x: np.percentile(x, 36) #своя функция, например 36 перцентиль
}).reset_index()
```

### Join / Union

```python
#перевернем данные в датафрейме и переименуем в нем названия полей, добавив _data_2

data_2 = data_2.reindex(index=data.index[::-1]).reset_index(drop=True)

data_2.columns = [str(c)+'_data_2' for c in data_2.columns]

data_2
```

```python
#джоин по умолчанию происходит по индексу
data.join(data_2)
```

```python
#вернем data_2 обратно
data_2 = data.copy()
```

```python
#более гибкий вариант
data.merge(data_2, how='left', on=['col_1', 'col_2'], suffixes=['_c_1', '_c_2'])
```

```python
#объединение датафреймов (аналог union в sql)
data.append(data_2)
```

### Оконные функции

```python
#row_number
data['animal_rank'] = data.sort_values(['col_2'], ascending=True
                                      ).groupby(['type']).cumcount() + 1

data[['type', 'col_2', 'animal_rank']]
```

```python
#сумма накопительным итогом
data['animal_cum_total_vol'] = data.sort_values(by=['dt'], ascending=True
                                               ).groupby(['type'])['some_value'].cumsum()

data[['type', 'some_value', 'animal_cum_total_vol']]
```

```python
import pandas as pd
import tinkoff as tf

query = """
    select toDate(event_dttm) day_dt, count(distinct mdm_party_rk) persons,
    count(distinct amplitude_session_id) sessions, count(program_product_event_id) events
    from  prod_v_usermart.program_product_event_extended
    where 
        event_dttm >= '{0}' and event_dttm <'{1}'
        and program_product_cd = 'MPL'
        and event_type_code = 'SME_Transfer_Done'
        and JSONExtractString(event_param_list, 'Screen') = 'main'
        group by toDate(event_dttm)"""

data_res = pd.DataFrame()

for i in range(0, 90, 3):
    start_val = pd.Timestamp('2022-02-01').date() + pd.Timedelta(days=i)
    second_val = start_val + pd.Timedelta(days=3)    
    i_df = tf.click_to_df(query.format(start_val, second_val), clickhouse_service='dwh')
    data_res = data_res.append(i_df)

tf.smb_upload(file_or_df=data_res, server_path='fs-nfsmb/zeppelin_exchange$/s.kostyuchik/data_res_real.csv',
              login='s.kostyuchik', password='')
```
