```python
df_properties = pd.read_csv('sme_mb_90_day_schema_3.csv')
```

```python
df_properties.columns = [c.replace(' ', '_').lower() for c in df_properties.columns]
```

```python
for row in df_properties[:100].itertuples():

    try:
        np.isnan(row.event)
    except TypeError:
        event_name = row.event
        event_desc = row.event_description
        num_props = row.event_number_of_properties
        
    print({
        'event_name':event_name, 'event_desc':event_desc,
        'num_props':num_props, 'prop_name':row.event_property,
        'prop_desc':row.property_description})
```

```python
parsed_data = pd.DataFrame(columns=['event_name', 'event_desc', 'num_props', 'prop_name', 'prop_desc'])

for row in df_properties.itertuples():

    try:
        np.isnan(row.event)
    except TypeError:
        event_name = row.event
        event_desc = row.event_description
        num_props = row.event_number_of_properties
        
    parsed_data = parsed_data.append({
        'event_name':event_name, 'event_desc':event_desc,
        'num_props':num_props, 'prop_name':row.event_property,
        'prop_desc':row.property_description},
        ignore_index=True)
```

```python
data = df[df['надо_переносить'].isin(['да', 'Да', 'д'])]
data = data[['event','event_description','event_number_of_properties']]
data = data.merge(parsed_data, left_on='event', right_on='event_name', how='left')
```

```python
data = data[['event','event_description','event_number_of_properties','prop_name','prop_desc']]

data.columns = ['event', 'event_disc', 'params_qty', 'param', 'param_disc']
```

```python
data.event_disc = data.event_disc.fillna('Необходимо добавить описание события')
data.param_disc = data.param_disc.fillna('Необходимо добавить описание параметра')
```

```python
data
```

```python
for event in tqdm(data.event.unique()):
    #print(event)
    
    iter_event_df = data[data['event']==event]

    e_disc = iter_event_df.event_disc[:1].squeeze()

    params_dict = {}

    if iter_event_df.params_qty[:1].squeeze() > 0:
        for p in iter_event_df.param.unique():
            p_disc = iter_event_df[iter_event_df['param']==p]['param_disc'].squeeze()

            params_dict.update({p: {'meta': {'description': p_disc},
                                    'alias': p, 'type': 'String', 'nullable': False}})

    event_file = {
        'meta': {'description': e_disc,
                 'maintainers': [],
                 'tasks': [],
                 'labels': ['sme_transfer_script']},
        'alias': {'system': 'amplitude',
                  'name': event},
        'parameters': params_dict,
        'settings': {},
        'clickhouse': {
            'enabled': True}

    }

    with open(f'sme_amplitude_events_3/{event}.yml', 'w') as file:
        try:
            yaml.dump(event_file, file, allow_unicode=True)
        except yaml.YAMLError as exc:
            print(exc)
```
